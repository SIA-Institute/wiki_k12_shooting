```python
import googlemaps
from datetime import datetime

gmaps = googlemaps.Client(key='YOUR API KEY')
```

## Documentation

[See link](https://googlemaps.github.io/google-maps-services-python/docs/index.html)

```python
def query_school(school_name):
    """ 
    Query by school_name and return its place_id if status shows 'OK' 
    """
    
    result = gmaps.find_place(school_name,input_type='textquery')
    if result['status'] == 'OK':
        return result['candidates'][0]['place_id']
    else:
        return pd.NA
```

```python
df['place_id'] = df['School_full_name'].apply(query_school)
```

```python
df.head()
```

```python
detail_dict = {}

for i, v in enumerate(df['place_id'].values):
    try:
        detail = gmaps.place(place_id=v)
        detail_dict[i] = detail
    except:
        detail_dict[i] = None
```

```python
def county_from_deatil(detail):
    for i in detail['result']['address_components']:
        if i['types'] == ['administrative_area_level_2', 'political']:
            return i['long_name']
```

```python
county_lst = []
for key, value in detail_dict.items():
    if value != None:
        county_lst.append(county_from_deatil(value))
    else:
        county_lst.append(pd.NA)
```

```python
df['County'] = county_lst
```
